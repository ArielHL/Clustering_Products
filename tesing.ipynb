#### Clustering by KMeans

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans

# Sample data setup (combine products from both Brand A and B)
products = [
    # Add your combined product list from both Brand A and B here
]

# Extract product names and ingredients as a single string per product for vectorization
data = [' '.join(product["Ingredients"]) for product in products]
product_names = [product["Product"] for product in products]

# Vectorizing the ingredients
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(data)

# Applying K-means clustering
kmeans = KMeans(n_clusters=5, random_state=0)  # You can adjust the number of clusters
kmeans.fit(X)

# Assigning the cluster labels to each product
clusters = kmeans.labels_

# Creating a DataFrame to view the products with their respective clusters
clustered_data = pd.DataFrame({
    'Product': product_names,
    'Cluster': clusters
})

# Printing the results
print(clustered_data.sort_values(by='Cluster'))
```

#### Using TF-IDF and HDBSCAN

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
import hdbscan

# Sample data setup (combine products from both Brand A and B)
products = [
    # Add your combined product list from both Brand A and Brand B here
]

# Extract product names and ingredients as a single string per product for vectorization
data = [' '.join(product["Ingredients"]) for product in products]
product_names = [product["Product"] for product in products]

# Vectorizing the ingredients with TF-IDF
tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(data)

# Dimensionality Reduction with PCA
pca = PCA(n_components=0.95)  # Adjust the number of components to capture 95% variance
X_pca = pca.fit_transform(X_tfidf.toarray())

# Clustering with HDBSCAN
clusterer = hdbscan.HDBSCAN(min_cluster_size=5, gen_min_span_tree=True)
cluster_labels = clusterer.fit_predict(X_pca)

# Creating a DataFrame to view the products with their respective clusters
clustered_data = pd.DataFrame({
    'Product': product_names,
    'Cluster': cluster_labels
})

# Printing the results
print(clustered_data.sort_values(by='Cluster'))
```

### Using Bertopic

```python
from bertopic import BERTopic
from sklearn.datasets import fetch_20newsgroups

# Sample data setup (assuming 'products' is already your dataset)
products = [
    # Add your combined product list from both Brand A and Brand B here
]

# Extract product names and a single string of ingredients per product for embedding
ingredients = [' '.join(product["Ingredients"]) for product in products]
product_names = [product["Product"] for product in products]

# Initialize BERTopic model
topic_model = BERTopic(language="english", calculate_probabilities=True, verbose=True)

# Fit the model and transform the data
topics, probabilities = topic_model.fit_transform(ingredients)

# Get an overview of the topics
topic_info = topic_model.get_topic_info()

# Retrieve individual topics and their representative ingredients
for topic_num in range(len(topic_info) - 1):  # Excluding the -1 outlier topic
    print(f"Topic {topic_num}: {topic_model.get_topic(topic_num)}")

# Mapping topics to products
topic_labels = [topics[idx] for idx in range(len(products))]
clustered_data = pd.DataFrame({
    'Product': product_names,
    'Topic': topic_labels
})

print(clustered_data.sort_values(by='Topic'))
```

#### User Transformers

```python
from transformers import AutoTokenizer, AutoModel
import torch
from sklearn.cluster import AgglomerativeClustering
import pandas as pd

# Sample data setup
products = [
    # Add your combined product list from both Brand A and Brand B here
]

# Preparing the ingredients text
ingredients_text = [' '.join(product["Ingredients"]) for product in products]
product_names = [product["Product"] for product in products]

# Load a pre-trained transformer model
tokenizer = AutoTokenizer.from_pretrained('distilbert-base-uncased')
model = AutoModel.from_pretrained('distilbert-base-uncased')

# Function to encode text to embeddings
def get_embeddings(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].detach().numpy()

# Convert ingredients to embeddings
embeddings = [get_embeddings(text) for text in ingredients_text]

# Clustering
clustering_model = AgglomerativeClustering(n_clusters=5)  # Adjust the number of clusters as needed
cluster_labels = clustering_model.fit_predict(torch.stack(embeddings))

# Organize the data into a DataFrame for visualization
clustered_data = pd.DataFrame({
    'Product': product_names,
    'Cluster': cluster_labels
})

print(clustered_data.sort_values(by='Cluster'))
```
