### Option Two using Kmeans

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd

# Combine the lists from both brands
products_data = products_brand_b + products_brand_a


# Step 1: Normalize ingredients
def normalize_ingredients(products):
    descriptions = []
    for product in products:
        # Create a string "ingredient/rank" for each ingredient
        description = ' '.join([f"{ing.split(':')[0]}/{round(1/float(ing.split(':')[1]),2)}" for ing in product['Ingredients']])
        descriptions.append(description)
    return descriptions

descriptions = normalize_ingredients(products_data)

# Step 2: Vectorize descriptions
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(descriptions)

# Step 3: Cluster the vectorized data
kmeans = KMeans(n_clusters=10, random_state=42, n_init=100, max_iter=1000)
clusters = kmeans.fit_predict(X)

# Step 4: Output the clusters
product_clusters = {}
for idx, cluster in enumerate(clusters):
    if cluster not in product_clusters:
        product_clusters[cluster] = []
    product_clusters[cluster].append(products_data[idx]['Product'])

product_clusters
clusters_df=pd.DataFrame(product_clusters.items(), columns=['Cluster', 'Products'])
```
